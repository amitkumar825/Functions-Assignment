```python
#10. What is the difference between map(), reduce(), and `filter() functions in Python?

Answer:-

#Map Functions:- The map function applies a given function to all items in an input iterable (like a list) and returns an iterator with the results.
Ex- l = [1, 2, 3, 4, 5]

    list(map(lambda x: x+10, l))
    Result- [11, 12, 13, 14, 15] 
    
    
#Reduce Functions:-  The reduce function from the functools module applies a given function cumulatively to the items of a sequence, from left to right, to reduce the sequence to a single value.
Ex- from functools import reduce

    words = ["data", "science", "courses"]
    reduce(lambda x, y: x+ " "+y, words)
    Result- 'data science courses'
    
    
#Filter Functions:- The filter function constructs an iterator from elements of an iterable for which a specified function returns True.
Ex- l = [2, 1, 3, 4, 4, 5, 6]

    list(filter(lambda x: x%2 == 0, l))
    Result- [2, 4, 4, 6]
```

```python
#5. Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms.

Answer:-

def fib(n):
    a = 0 
    b = 1
    for i in range(n): 
        yield a
        a, b = b, a+b
        
f = fib(100000)
next(f)
Results- 0

next(f)
Results- 1

next(f)
Results- 1

next(f)
Results- 2

next(f)
Results- 3

next(f)
Results- 5
```

```python
#6. Write a generator function in Python that yields the powers of 2 up to a given exponent.

Answer:-

def square_number_generators(n):
    for i in range(n):
        yield i ** 2

gen = square_number_generators(5)
gen
Results- <generator object square_number_generators at 0x7d08c8187c30>

next(gen)
Results- 0

next(gen)
Results- 1

next(gen)
Results- 4

next(gen)
Results- 9

next(gen)
Results- 16
```

```python
#9. Write a Python program that uses 'map()` to convert a list of temperatures from Celsius to Fahrenheit.

Answer:-

celsius_temperatures = [0, 20, 37, 100, -10]

def celsius_to_fahrenheit(celsius):
    return (celsius * 9/5) + 32
fahrenheit_temperatures = map(celsius_to_fahrenheit, celsius_temperatures)
fahrenheit_temperatures_list = list(fahrenheit_temperatures)
print(fahrenheit_temperatures_list)
Results- [32.0, 68.0, 98.6, 212.0, 14.0]
```

```python
#10. Create a Python program that uses 'filter()` to remove all the vowels from a given string.

Answer:-

def is_not_vowel(char):
    return char.lower() not in 'aeiou'

def remove_vowels(input_string):
    filtered_chars = filter(is_not_vowel, input_string)
    return ''.join(filtered_chars)

remove_vowels("pwskills")
Results- 'pwsklls'

remove_vowels("amit")
Results- 'mt'

remove_vowels(" i am going to delhi")
Results- '  m gng t dlh'
```

```python
#11) Imagine an accounting routine used in a book shop. It works on a list with sublists, which look like this:

order Number       Book Title and Author                               Quantity                     Price per Item
34587              Learning Python, Mark Lutz                             4                            40.95
98762              Programming Python, Mark Lutz                          5                            56.80
77226              Head First Python, Paul Barry                          3                            32.95
88112              Einführung in Python3, Bernd Klein34587                3                            24.99

Write a Python program, which returns a list with 2-tuples. Each tuple consists of the order number and the product of the price per item and the quantity. The product should be increased by 10,- € if the value of the order is smaller than 100,00 €.

Write a Python program using lambda and map.


Answer:-

data = [
    [34587, "Learning Python, Mark Lutz", 4, 40.95],
    [98762, "Programming Python, Mark Lutz", 5, 56.80],
    [77226, "Head First Python, Paul Barry", 3, 32.95],
    [88112, "Einführung in Python3, Bernd Klein", 3, 24.99]
]
calculate_total = lambda order: (
    order[0],  
    (order[2] * order[3]) + (10 if (order[2] * order[3]) < 100 else 0) 
)
result = list(map(calculate_total, data))

print(result)
Results- [(34587, 163.8), (98762, 284.0), (77226, 108.85000000000001), (88112, 84.97)]
```
